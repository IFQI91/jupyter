### Mi primera base de datos

```sas
DATA uno;
input x y nombre $;
datalines;
1.87 75 juan
2.0 85 luis
1.5 60 alex
1.67 68 fer
;
run;
proc print data=uno;
run;
```

```sas
proc means; run;
```

```sas
proc sgscatter data=uno;
plot y*x
/datalabel= nombre group= nombre;
run;
```
