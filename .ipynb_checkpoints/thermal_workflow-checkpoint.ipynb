```python
%matplotlib inline
import os
import numpy as np
from plantcv import plantcv as pcv
import plantcv.parallel
import plantcv.utils
```

```python
# Handle input variables
# This code was provided to make it easier to convert the notebook to a script
class options():
    def __init__(self):
        self.tmp_dir = "json_files"
        self.debug = "plot"
        self.result = os.path.join(self.tmp_dir, "0001_IR.txt")
        self.outdir = "./output_images"
        self.writeimg = False
        self.image = "0001_IR.csv"
        os.makedirs(self.tmp_dir, exist_ok=True)
```

```python
# Read input options
args = options()
pcv.params.debug = args.debug
pcv.params.dpi = 150
```

```python
# Read image file in "CSV" mode
img, imgpath, imgname = pcv.readimage(filename=args.image, mode="csv")
```

```python
min_temp = np.min(img)
max_temp = np.max(img)
print(f"Min temp: {min_temp}")
print(f"Max temp: {max_temp}")
```

```python
# Create a mask that keeps the whole image unless there is a subregion of interest
mask = np.ones(img.shape, dtype=np.uint8)
```

```python
thermal_hist = pcv.analyze_thermal_values(thermal_array=img, mask=mask, histplot=True)
```

```python
ps_img = pcv.visualize.pseudocolor(gray_img=img, min_value=int(min_temp + 0.5), max_value=int(max_temp + 0.5), cmap="plasma")
```

```python
pcv.print_results(filename=args.result)
```

```python
plantcv.parallel.process_results(job_dir=args.tmp_dir, json_file="0001_IR.json")
plantcv.utils.json2csv(json_file="0001_IR.json", csv_file="0001_IR")
```
